```python
import pandas as pd

dataset = pd.read_csv("./data/listings_barcelona.csv")
dataset.head(5)
```

```python
df = dataset.copy()
```

```python
df.minimum_nights
df.host_since
```

```python
import numpy as np

df['host_since'] = pd.to_datetime(df['host_since']) #cast to dates
df['last_review'] = pd.to_datetime(df['last_review'])#cast to dates
df['price'] = df['price'].replace("[$,]", "", regex=True).astype(float) #cast text as currency
#df['minimum_nights'] = df['minimum_nights'].astype('int')


host_experience_limit = np.datetime64('2023-09-01') # One year as host at least(at September 2024)
reviews_limit = np.datetime64('2024-03-01') # six months limit
minimum_nights_limit = 10


df = df[df['host_since'] <= host_experience_limit] #host with one year experience or more only
df = df[df['last_review'] >= reviews_limit] # six months reviewed only
df = df[df['minimum_nights'] <= minimum_nights_limit]
df['minimum_nights']
```

```python
reviewed_df = df.dropna(subset=['last_review','reviews_per_month', 'price'], how='all', inplace=False) #delete all rows without reviews per month or a last review data
reviewed_df
```

```python
accepted_types = ['Entire home/apt','Private room']
reviewed_df = reviewed_df[reviewed_df['room_type'].isin(accepted_types)] # only accepted types
reviewed_df
```
